```python

def sortRowWise(m):
    p=m
    # loop for rows of matrix
    for i in range(len(p)):
         
        # loop for column of matrix
        for j in range(len(p[i])):
             
            # loop for comparison and swapping
            
            for k in range(len(p[i]) - j - 1):
                 
                if (p[i][k] > p[i][k + 1] and k>0):
                     
                    # swapping of elements
                    t = p[i][k]
                    p[i][k] = p[i][k + 1]
                    p[i][k + 1] = t
    return p
```

```python
def top5(f):
    top5=[0]*d
    sortRowWise(f)
    for i in range(d):
        top5[i]=f[i][c]+f[i][c-1]+f[i][c-2]+f[i][c-3]+f[i][c-4]
    return top5
    
```
